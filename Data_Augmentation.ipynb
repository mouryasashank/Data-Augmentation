```
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from cv2 import resize
from skimage import exposure, transform
from skimage.util import random_noise
```

```
image= mpimg.imread('/content/da4.jpg')
plt.imshow(image)
```

```
image_rescale= resize(image,(500,500))
plt.imshow(image_rescale)
```

```
horflip= np.fliplr(image_rescale)
plt.imshow(horflip)
```

```
verflip= np.flipud(image_rescale)
plt.imshow(verflip)
```

```
trans= transform.rotate(image_rescale, random.uniform(-30,30))
plt.imshow(trans) 
```

```
noise= random_noise(image_rescale, mode='s&p', clip=True)
plt.imshow(noise)
```
